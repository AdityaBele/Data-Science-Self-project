## 3 : Function use to find missing Age Value

    By applying Function model and predection from boxplot.

```python
def ipmute_age(cols):
    Age = cols[0]
    Pclass= cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 28
        else:
            return 23
    else:
        return Age
        
    
    
```

Now Apply the Function

```python
train['Age']=train[['Age','Pclass']].apply(ipmute_age,axis=1)
```
