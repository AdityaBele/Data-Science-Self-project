# Spotify Data Analysis using Python

```python
import pandas as pd 
import numpy as np
```

## Data Collection

```python
df = pd.read_csv('spotifydataset.csv')
```

```python
df
```

```python
df.info()
```

## Data Wrangling or Cleaning 

```python
df2 = df.dropna()
```

```python
df2.info()
```

```python
df2.describe().transpose()
```

```python
most_pop = df2.query('popularity>90', inplace = False).sort_values('popularity',ascending=False)
```

```python
most_pop[:20]
```

```python
df2['duration']= df2['duration_ms'].apply(lambda x:round(x/1000))
df2.drop("duration_ms",inplace = True,axis=1)
```

```python
df2
```

```python
df2.duration.head()
```

## Data Visulation 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
corr_df = df2.drop(['key','mode','explicit','track_id','artists','album_name','track_name','track_genre','Unnamed: 0'],axis = 1).corr(method="pearson")
plt.figure(figsize=(14,6))
heat = sns.heatmap(corr_df,annot=True,fmt ='.1g',vmax = 1, vmin = -1,center = 0,cmap="inferno")
heat.set_title("correlation Heatmap btw variable")
```

```python
df2_sam = df2.sample(int(0.04*len(df2)))
```

```python
print(len(df2_sam))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = df2_sam , x = 'loudness',y = 'energy',color = 'b').set(title = 'Loudness vs Energy Correlation')
```

```python
plt.figure(figsize=(18,20))
plt.title('Duration of Songs in diffrent Genre')
sns.color_palette('rocket', as_cmap = True)
sns.barplot(data = df2 , x = 'duration',y = 'track_genre')
plt.xlabel('Duration of song in sec')
plt.ylabel('Genre of the Song')
```

```python
sns.set_style(style = 'darkgrid')
plt.figure(figsize=(15,8))
famous = df2.sort_values('popularity', ascending = False).head(10)
sns.barplot(y='track_genre',x = 'popularity',data = famous).set(title = 'Top 5 & more populare Genre')
```
